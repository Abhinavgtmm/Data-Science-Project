```python
Gender = pd.get_dummies(auto['Gender'], drop_first = True)
Gender
```

```python
auto = pd.concat([auto, Gender], axis = 1)
auto
```

```python
len(set(auto['Product Category']))
```

```python
auto = auto.drop(['Gender','Product Category'],axis = 1)
auto
```

```python
auto['Date'] = pd.to_datetime(auto['Date'])
auto['DayOfWeek'] = auto['Date'].dt.dayofweek
auto = auto.drop('Date', axis=1)
auto.corr()
```

```python
# Starting machine learning
import seaborn as sns
plt.figure(figsize=(15,5))
sns.heatmap(auto.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(auto)
```

```python
auto = auto.drop(['Male'],axis = 1)
auto
```

```python
auto['Total Amount'] = auto['Total Amount'].astype('int')
```

```python
from sklearn.model_selection import train_test_split
# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(auto, train_size = 0.85, test_size = 0.15, random_state = 1)
```

```python
df_train
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
from sklearn.linear_model import LinearRegression
X_train = df_train[['Age',	'Quantity',	'Price per Unit',	'DayOfWeek']]

y_train = df_train['Total Amount']
# df_test_predict = X_train.iloc[0:1]

# for linear regression
lr = LinearRegression()
lr_model = lr.fit(X_train, y_train)

# # #for logistic regression
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg_model = lg.fit(X_train, y_train)

# # # # #for polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=8)
X_poly = poly_reg.fit_transform(X_train)
# Y_poly = poly_reg.fit_transform(y_train)
pol_reg = LinearRegression()
model = pol_reg.fit(X_poly, y_train)
poly_reg.fit(X_train, y_train)
```

```python
# #to dump model
import pickle as pk
filename = 'model.pickle'
pk.dump(lr_model, open(filename, 'wb'))
```

```python
data = df_test.iloc[50:51]
actual_totalamount = data['Total Amount']
data = data.drop(['Total Amount'],axis = 1)
data
```

```python
#linear model
print("Predicted Total Amount	",lr_model.predict(data))
print("Actual Total Amount	",actual_totalamount)
```

```python
# #logistic
print("Predicted Total Amount",lg_model.predict(data))
print("Actual Total Amount",actual_totalamount)
```

```python
# # #polynomial regression model test
predicted_totalamount = pol_reg.predict(poly_reg.fit_transform(data))
print("Actual Total Amount",actual_totalamount)
print("predicted Total Amount",predicted_totalamount)
```

```python
#train acc
print("Linear regression = ",round(lr_model.score(X_train, y_train)*100))
print("Logistic regression = ",lg_model.score(X_train, y_train)*100),2
print("Polynomial regression = ",pol_reg.score(poly_reg.fit_transform(X_train), y_train)*100)

```

```python
#test acc
test_data = df_test
y_test = test_data['Total Amount']
X_test = test_data.drop(['Total Amount'],axis = 1)

print("Linear regression = ",round(lr_model.score(X_test, y_test)*100))
print("Logistic regression = ",lg_model.score(X_test, y_test)*100)
print("Polynomial regression = ",pol_reg.score(poly_reg.fit_transform(X_test), y_test)*100)

```
